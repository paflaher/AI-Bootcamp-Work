```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

raw_data = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv') 

print(raw_data.head())
```

```python
raw_data.columns
```

```python
raw_data['sepal_length']
```

```python
raw_data.replace('nan', np.nan, inplace=True)
for col in raw_data.columns:
    print(col, raw_data[col].isnull().sum())
```

```python
# create a copy of this data set to work with 

iris = raw_data.copy()
```

```python
# check that the copy worked

iris.head()
```

```python
iris.describe(include='all')
```

```python
max_sepal_width = iris['sepal_width'].max()
min_petal_length = iris['petal_length'].min()

print("Max Sepal Width:", max_sepal_width)
print("Min Petal Length:", min_petal_length)
```

```python
species_count = iris['species'].value_counts()
print(species_count)
```

```python
# This will give the average sepal length and petal length

mean_sepal_length = iris.groupby('species')['sepal_length'].mean() # Gets the mean length of the sepal for each species and saves to a variable

mean_petal_length = iris.groupby('species')['petal_length'].mean() # Gets the mean length of the petal for each species and saves to a variable

print(mean_sepal_length)
print(mean_petal_length)
```

```python
# Bar graph to show the mean sepal lenght of each species 

mean_sepal_length.plot(kind='bar', title='Mean Sepal Length for Each Species of Iris') # Plots the mean sepal length for each species of iris as a type of bar graph
plt.xlabel('Species') # Label for the x axis
plt.ylabel('Mean Sepal Length (cm)') # Label for the y axis
plt.show() # Shows the plot
```

```python
# pie chart to show the mean petal length of each species

mean_petal_length.plot(kind='pie', title='Mean Petal Length for Each Species of Iris') # Plots the mean petal length for each species of iris as a type of pie chart
plt.ylabel('')
plt.show()
```

```python
# scatter plot showing the relationship between sepal length and sepal width

iris.plot(kind='scatter', x='sepal_length', y='sepal_width', title='Sepal Length vs Sepal Width') # Plots the sepal length against the sepal width as a scatter plot
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# a hist chart showing the relationship between petal length and petal width

iris.plot(kind='hist', x='petal_width', y='petal_length', rwidth=0.6, title='Petal Width vs Petal Length')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# box plot of the sepal length of each species

iris.boxplot(column='sepal_length', by='species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.title('')
plt.xticks(rotation=90)
plt.show()
```
